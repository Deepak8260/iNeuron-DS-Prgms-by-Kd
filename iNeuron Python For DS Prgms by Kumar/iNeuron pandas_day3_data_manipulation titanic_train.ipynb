```python

df17=pd.DataFrame(np.array([[1,34,43,56],[5,62,73,84],[9,10,71,172]]),columns=['a','b','c','d'],index=[0,1,2])
df18=pd.DataFrame(np.array([[1,67,89,41],[5,76,67,85],[9,170,11,162]]),columns=['a','b','c','d'],index=[3,4,5])
```

```python
df17
```

```python
df18
```

```python
pd.merge(df17,df18,on='a') #merging of two df on the basis of column a as column a has same values. by default it perform inner merge, df17 intersection df18
```

```python
df19=pd.DataFrame(np.array([['k0','a0','b0'],['k8','a1','b1'],['k2','a2','b2'],['k3','a3','b3']]),columns=['key','a','b'],index=[0,1,2,3])
df20=pd.DataFrame(np.array([['k0','c0','d0'],['k1','c1','d1'],['k2','c2','d2'],['k3','c3','d3']]),columns=['key','a','b'],index=[0,1,2,3])
```

```python
df19
```

```python
df20
```

```python
pd.merge(df19,df20,on='key',how='outer')#outer merge .merge operation is performed on the basis of column
```

```python
pd.merge(df19,df20,on='key',how='right') #right merge
```

```python
df21=pd.DataFrame(np.array([['k0','a0','b0'],['k8','a1','b1'],['k2','a2','b2'],['k3','a3','b3']]),columns=['key','a','b'],index=[0,1,2,3])
df22=pd.DataFrame(np.array([['k0','c0','d0'],['k1','c1','d1'],['k2','c2','d2'],['k3','c3','d3']]),columns=['key','c','d'],index=[0,1,2,3])
```

```python
df21
```

```python
df22
```

```python
pd.merge(df21,df22,on='key',how='outer') #outer merge = left +  right merge
```

```python
df23=pd.DataFrame(np.array([['a0','b0'],['a1','b1'],['a2','b2']]),columns=['a','b'],index=['k0','k1','k2'])
df24=pd.DataFrame(np.array([['c0','d0'],['c2','d3'],['c3','d3']]),columns=['c','d'],index=['k0','k2','k3'])
```

```python
df23
```

```python
df24
```

```python
df23.join(df24) #left join by default. Same operations we can do as we can in merge
```

```python
df23.join(df24,how='right') # in join , we perform the operation on the basis of index or rows but merge operation is performed on the basis of column
```

```python
df23.join(df24,how='outer') 
```

```python
df23.join(df24,how='inner') 
```
