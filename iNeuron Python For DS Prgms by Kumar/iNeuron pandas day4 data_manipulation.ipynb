```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(20)
df=pd.DataFrame(np.random.randn(10,4),index=pd.date_range('12-07-2024',periods=10),columns=['a','b','c','d'])
df
```

```python
df['new']=df['c']+df['d']
```

```python
df
```

```python
df['new_column_sq_d'] = df['d'].apply(lambda x: x**2)
```

```python
df
```

```python
df['new_column_log_d'] = df['d'].apply(lambda x: np.log(x))
df # as applying log to -ve values, we get NaN values, so we need to filter the -ve values( see next cell)
```

```python
def test(x):
    if (x<0):
        return x
    else:
        return np.log(x)
```

```python
df['new_column_new_log_d'] = df['d'].apply(test)
df #here it keeps the same value of d when d is -ve and find the log value of +ve values
```

```python
def test1(x):
    if x<0:
        return 1
    else:
        return 0
```

```python
df['new_column_new_a'] = df['a'].apply(test1)
```
