```python
import scipy.stats as stats
x=[1,2,1,2,4,5,8,10,11,2,1,3,4,5,2,110,111]
xs=pd.DataFrame(x)
xs.iplot(kind='box')#showing the outlier with the help of plotly
```

```python
sns.boxplot(data=x) #showing the outlier with the help of seaborn
```

```python
plt.boxplot([x]) # #showing the outlier with the help of matplotlib
plt.show()
```
