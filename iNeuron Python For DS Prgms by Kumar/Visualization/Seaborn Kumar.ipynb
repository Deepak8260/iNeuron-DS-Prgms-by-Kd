```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='tip',y='total_bill',data=tips)
```

```python
sns.relplot(x='tip',y='total_bill',data=tips,hue = 'smoker') #hue means color. Orange color are non smoker and blue are the smokers.
```

```python
#sex wise graph
sns.relplot(x='tip',y='total_bill',data=tips,hue = 'sex')
```

```python
#day wise graph
sns.relplot(x='tip',y='total_bill',data=tips,hue = 'day') # here we cant say that in thursday, who are the smoker and who are not
```

```python
sns.relplot(x='tip',y='total_bill',data=tips,hue = 'day',style='smoker')# but here we can sapecify that in thursday, who are the smoker
```

```python
sns.relplot(x='tip',y='total_bill',data=tips,hue = 'day',style='smoker',size='size')
```

```python
sns.relplot(x='tip',y='total_bill',data=tips,size='size')
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y, marker='o')
# Add titles and labels
plt.title("Sample Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```
