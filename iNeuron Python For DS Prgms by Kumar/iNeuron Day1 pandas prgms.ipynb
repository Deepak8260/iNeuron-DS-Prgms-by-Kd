```python
#task is to collect user data from each list and display the df
for i in range(len(js)):
    print(pd.DataFrame([js[i]['user']])) #this code collects user data individually from each list and prints individually
```

```python
list=[]
for i in range(len(js)):
    list.append((js[i]['user']))
pd.DataFrame(list) #this approach 1st collects all the user data and store it into and empty list and then finally convert that list into df
```

```python
df13 = pd.DataFrame([js[i]['user'] for i in range(len(js))]) # same thing done but using list comprehension
```

```python
df14 = pd.DataFrame.from_records(js, columns=['user']) Smae thing 
df14
```
